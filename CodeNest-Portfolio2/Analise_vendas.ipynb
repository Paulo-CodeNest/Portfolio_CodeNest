```python
import pandas as pd

df_coisa = pd.read_csv("dados/bu_mz.csv")
display(df_coisa)
```

```python
total_vendas = len(df_coisa)
print("Total de vendas: ", total_vendas)
display(list(df_coisa.columns))
display(df_coisa.head())
```

```python
df_coisa["Total"] = df_coisa["Preco"] * df_coisa["Quantidade"]
display(df_coisa)
```

```python
soma_coluna = df_coisa["Total"].sum()
media_total = df_coisa["Total"].median()

print("Total :", soma_coluna, "Media: ", media_total)
```

```python
analise = df_coisa[["Provincia", "Total"]].groupby("Provincia").sum()
analise = analise.sort_values(by="Total", ascending=False)

display(analise)
```

```python
analise_1 = df_coisa[["Produto", "Total"]].groupby("Produto").sum()
analise_1 = analise_1.sort_values(by="Total", ascending=False)

display(analise_1)
analise_total = analise_1["Total"].sum()
print("O total é: ", analise_total)
```

```python
df_resumo = pd.DataFrame({
    "Item": ["Soma do total de vendas", "Media do total de vendas", "Total de vendas"],
    "Resultado": [ soma_coluna, media_total, total_vendas]
})
```

```python
from fpdf import FPDF
import os

def gerar_pdf_resumo(df_resumo, nome_arquivo, titulo="Projecto Portfolio - CodeNest"):
    pdf = FPDF()
    pdf.add_page()
    
    # Título
    pdf.set_font("Arial", "B", 16)
    pdf.cell(0, 10, titulo, ln=True, align="C")
    pdf.ln(10)

    # Corpo do resumo
    pdf.set_font("Arial", "", 12)
    
    for index, row in df_resumo.iterrows():
        linha = f"{row['Item']}: {row['Resultado']}"
        pdf.multi_cell(0, 8, linha)
        pdf.ln(2)

    caminho = os.path.join(os.getcwd(), nome_arquivo)
    pdf.output(caminho)

    print(f"PDF criado com sucesso em: {caminho}")

# Gerar PDF
gerar_pdf_resumo(df_resumo, "outputs/Resumo_CodeNest.pdf")
```
