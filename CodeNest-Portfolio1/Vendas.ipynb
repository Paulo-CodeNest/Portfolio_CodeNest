```python
import pandas as pd
from fpdf import FPDF
from tabulate import tabulate

df_vendas = pd.read_excel("dados/Dados_Organizados_Vendas_CodeNest.xlsx", engine="openpyxl")
display(df_vendas)
```

```python
produto_mais_vendido = df_vendas.groupby(["Produto"])["Quantidade"].sum()
produto_mais_vendido = produto_mais_vendido.idxmax()
print("Produto mais vendido: ", produto_mais_vendido)
```

```python
produto_mais_rentavel = df_vendas.groupby(["Produto"])["Total"].sum()
produto_mais_rentavel = produto_mais_rentavel.idxmax()
print("O produto que mis gerou dinheiro eh: ", produto_mais_rentavel)
```

```python
provincia_com_mais_vendas = df_vendas.groupby(["Provincia"])["Quantidade"].sum()
provincia_com_mais_vendas = provincia_com_mais_vendas.idxmax()
print("Provincia com mais vendas: ", provincia_com_mais_vendas)
```

```python
cliente_que_mais_comprou = df_vendas.groupby(["Cliente"])["Quantidade"].sum()
cliente_que_mais_comprou = cliente_que_mais_comprou.idxmax()
print("Cliente que mais comprou: ", cliente_que_mais_comprou)
```

```python
receita_total = df_vendas["Total"].sum()
print("A receita total é : ", receita_total, "MZN")
```

```python

# Criar DataFrame resumo
df_resumo = pd.DataFrame({
    "Item": ["Produto mais vendido", "Produto mais rentavel", "Província com mais vendas", "Cliente que mais comprou", "Receita Total"],
    "Resultado": [produto_mais_vendido, produto_mais_rentavel, provincia_com_mais_vendas, cliente_que_mais_comprou, receita_total]
})
```

```python
from fpdf import FPDF
import os

def gerar_pdf_resumo(df_resumo, nome_arquivo, titulo="Resumo Executivo - CodeNest"):
    pdf = FPDF()
    pdf.add_page()
    
    # Título
    pdf.set_font("Arial", "B", 16)
    pdf.cell(0, 10, titulo, ln=True, align="C")
    pdf.ln(10)

    # Corpo do resumo
    pdf.set_font("Arial", "", 12)
    
    for index, row in df_resumo.iterrows():
        linha = f"{row['Item']}: {row['Resultado']}"
        pdf.multi_cell(0, 8, linha)
        pdf.ln(2)

    caminho = os.path.join(os.getcwd(), nome_arquivo)
    pdf.output(caminho)

    print(f"PDF criado com sucesso em: {caminho}")

# Gerar PDF
gerar_pdf_resumo(df_resumo, "outputs/Resumo_Vendas_CodeNest.pdf")
```
